***주사위 빈도 계산하기***

- 주사위를 10000번 던져서 나오는 값들도 빈도를 계산하는 프로그램을 작성해보자. (for문으로 10000번 반복)

- 즉, 1,2,3,4,5,6이 몇 번 나오는지 계산한다. (if문 사용하여 +1)

- 난수 발생 함수(1~6 사이의 랜덤 숫자를 얻어야 함)와 리스트(각 숫자를 카운트하여 리스트에 저장)를 사용한다.

```python
import random

#리스트 초기화
dice_cnt = [0,0,0,0,0,0]

#10000번 반복
for i in range(10000):

#1~6 랜덤숫자 발생
    number = random.randint(1,6)

#횟수 카운트
    if number == 1:
            dice_cnt[0]+=1
    elif number == 2:
            dice_cnt[1]+=1
    elif number == 3:
            dice_cnt[2]+=1
    elif number == 4:
            dice_cnt[3]+=1
    elif number == 5:
            dice_cnt[4]+=1
    elif number == 6:
            dice_cnt[5]+=1
#결과 출력
print(dice_cnt)
```

과목별 점수의 평균 구하기
[[]]
- [국어, 영어, 수학]점수가 저장된 리스트이다.
- 각 과목의 리스트를 분리하고, 과목별 평균을 구해보자.
- 평균은 소수점 1자리까지 출력한다. 
    round(sum(koli)/len(koli),1)


```python
scores = [[96,84,80], [96,86,76], [76,95,83], [89,96,69], [90,76,91], [82,66,88], [83,86,79], [85,90,83]]
koli = []
enli = []
mali = []

for score in scores: 
    koli.append(score[0]) 
    enli.append(score[1])
    mali.append(score[2])

print(koli)
print(enli)
print(mali)

ave1 = round(sum(koli)/len(koli),1)
ave2 = round(sum(enli)/len(enli),1)
ave3 = round(sum(mali)/len(mali),1)
print(f'국어평균:{ave1} ,영어평균: {ave2}, 수학평균: {ave3} ')
```

튜플(Tuple)

- 튜플은 아래의 두가지 점을 제외하곤 리스트와 거의 동일하다.
    - 리스트는 [], 튜플은 () 사용
    - 리스트는 값의 생성, 삭제, 수정이 가능하지만 튜플은 값을 바꿀 수 없다.

- 튜플을 언제 사용할까?
    - 프로그램이 실행되는 동안 값이 바뀌면 안되는 경우

- 튜플 생성하기
    - 튜플명 = (요소1, 요소2,..., 요소N)
    - 튜플명 = 요소1, 요소2,..., 요소N
    - 요소가 한 개인 경우
        - 튜플명 = (요소,)
        - 튜플명 = 요소,

```python
#튜플 다루기

t1 = (1,5,3,4,2)
t2 = (6,7,8,9,8)

print(t1[1]) #인덱싱
print(t1[2:4]) #슬라이싱
print(t1+t2) #더하기
print(t1*3)#곱하기
print(len(t1)) #길이 구하기
print(t2.index(9)) #요소의 위치 찾기
print(t2.count(8)) #요소의 개수

#정렬된 새로운 리스트 반환
sort_t1 = sorted(t1)
print(sort_t1) #리스트의 형태로 반환
```

예제2> 단어장 만들기

- 엔터(")를 입력할 때까지 영어단어-뜻을 입력받아 단어장을 만들고(while문을 사용하여 반복, 딕셔너리 사용), 입력이 끝나면 단어테스트를 실시하는 프로그램을 만들어보자.

- 리스트에 입력된 순서대로 테스트를 실시하고( for문으로 딕셔너리 전체를 순회), 맞은 개수를 계산한다(맞으면 1씩 더하기).

- 테스트가 끝나면 맞은 점수를 출력한다.

- 출력형식 = 맞은개수/전체단어수 : 6/8

```python
#엔터(")를 입력할 떄 까지 영어단어-뜻을 입력받아 단어장을 만든다 (오류 뜸, 문제 원인 찾기**)

dict_voca = {}

while True:
    eng = input('eng: ')
    if eng == '':
        break #값을 입력하지 않고 엔터를 입력할 떄 까지
    kor = input('kor: ')
    dict_voca[eng] = kor
dict_voca

#단어테스트를 실시한다. 
#리스트에 입력된 순서대로 테스트를 실시하고 맞은 개수를 계산한다.

score = 0 
for english in dict_voca.keys():
    kor in input(english + ':')
    if kor == dict_voca[english]:
        score += 1
print(score)
```
